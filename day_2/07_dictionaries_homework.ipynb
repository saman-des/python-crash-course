TASK 4: Country Capital Look-up
Create a dictionary called 'capitals' with at least 5 countries and their capitals.
Example: "Nepal": "Kathmandu"
Ask the user to type a country name.
If it is in the dictionary, print its capital.
If not, print "Country not found!"
Write your code below:

```python
capitals = {
    "Nepal": "Kathmandu",
    "France": "Paris",
    "Japan": "Tokyo",
    "Brazil": "Brasilia",
    "Australia": "Canberra"
}

user_input = input("Enter a country name to find its capital: ")

if user_input in capitals:
    print(f"Capital is {capitals[user_input]}")
else:
    print("Not found")
```
