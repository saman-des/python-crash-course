TASK 3: Number List Maths
Create a list called 'test_scores' with 5 numbers (your choice).
Print:
  - The highest score  (use max())
  - The lowest score   (use min())
  - The total          (use sum())
  - Scores sorted low → high  (use sorted())
Write your code below:

```python
test_scores = [85, 92, 78, 95, 88]

print(f"Highest score: {max(test_scores)}")

print(f"Lowest score:  {min(test_scores)}")

print(f"Total points:  {sum(test_scores)}")

print(f"Scores (Sorted): {sorted(test_scores)}")
```
