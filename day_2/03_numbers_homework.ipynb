TASK 4: Price Calculator
Ask the user:
  - How many items they want to buy  (use int())
  - The price per item               (use float())
Calculate the total cost and print it.
Write your code below:

```python
quantity = int(input("How many items you wanna buy?"))

price_per_item = float(input("What is the price per item? "))

total_cost = quantity * price_per_item

print(f"Total cost for {quantity} items: ${total_cost:.2f}")
```
